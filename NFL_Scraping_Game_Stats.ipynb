```python
def add_opponents(df, assign_dict):
    
    opp_list = [] 
    for index, row in df.iterrows():
        team = row['team']
        week = row['week']
                
        game_week = assign_dict[week]
        
        opp_list.append(game_week[team])
    
    return opp_list
    
opponents = add_opponents(rb_weekly, weekly_opponents)
rb_weekly 
```
