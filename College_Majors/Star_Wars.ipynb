# Analyzing Star Wars movies
In this project, I will be analyzing reviews on Star Wars movies. It's a fun little side project that I worked on while waiting for Star Wars: The Rise of Skywalker. But I only had access to the first 6 episodes in the Saga not including the other 2 episodes. It has been said by many star wars fans that Episode 5 is currently the best star wars movie of all-time. I'll be exploring that.

```python
#Reading the Star Wars csv file I had access to using pandas.
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
pd.set_option('display.max_columns', None)
star_wars.head(10)
```

```python
#Finding the amount of rows and columns the dataset has.
star_wars.shape

#It has 1187 rows and 38 columns.
```

```python
#Familiarizing myself with the names of the columns before continuing.
star_wars.columns
```

```python
#Removing any rows where RespondentID is NaN.
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head()
```

```python
star_wars.shape

#Only 1 row was removed.
```

```python
#Converting the following series below to Boolean types.
yes_no = {
    "Yes": True,
    "No": False
}

#Changing the values in the column from "Yes" or "No" to Boolean values.
columns = ["Have you seen any of the 6 films in the Star Wars franchise?",
          "Do you consider yourself to be a fan of the Star Wars film franchise?"]

for col in columns:
    star_wars[col] = star_wars[col].map(yes_no)
```

```python
#Checking that the values of the series above only contain the values: True, False, and NaN/.
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna=False)
```

```python
#Checking that the values of the series above only contain the values: True, False, and NaN/.
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna=False)
```

```python
import numpy as np

#Converting column values to Boolean.
Episodes_seen_unseen = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.NaN: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi":True
}
        
#Converting all the values in the columns above to boolean values.
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(Episodes_seen_unseen)

star_wars.head(10)
```

```python
#Renaming columns.
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1", 
                                     "Unnamed: 4":"seen_2",
                                     "Unnamed: 5":"seen_3",
                                     "Unnamed: 6":"seen_4",
                                     "Unnamed: 7":"seen_5",
                                     "Unnamed: 8":"seen_6"})
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1"
    , "Unnamed: 10":"ranking_2", 
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
})

star_wars.head(10)
```

```python
#Generating a barplot to visualize the rankings between each of the movies.
import matplotlib.pyplot as plt
%matplotlib inline

rank_means = star_wars[star_wars.columns[9:15]].mean()
fig,ax = plt.subplots()
ax.bar(range(6), rank_means, color='r', width=0.25)
ax.set_title("Star Wars Episodes Fan Rankings")
ax.set_ylabel("Rankings")
ax.set_xlabel("Star Wars Episodes")
ax.set_xticklabels(("Episode 1", "Episode 2", "Episode 3", "Episode 4", "Episode 5", "Episode 6"))
```

What I've done here is that I have created a bar graph that visualizes the average rankings for each of the original 6 star wars movies in the saga. According to this visual, Episode 5 does have the highest average fan ranking (the lower the bar, the better). However, this visual isn't concrete enough. There are still other things we have to consider when coming up with our final answer. 

```python
#Computing the Sum of each of the seen columns.
seen_sums = star_wars[star_wars.columns[3:9]].sum()
```

```python
#Creating a bar chart for the sums in the seen columns.
fig, ax = plt.subplots()
ax.bar(range(6), seen_sums, width=0.25)
ax.set_title("Star Wars Episodes Seen")
ax.set_ylabel("Rankings")
ax.set_xlabel("Star Wars Episodes")
ax.set_xticklabels(("Episode 1", "Episode 2", "Episode 3"
                    , "Episode 4", "Episode 5", "Episode 6"))
```

What I've done here is I've created another visual that illustrates the number of people in the poll that have actually seen the star wars movies. What's interesting is that most people in the poll have actually seen Episode 5. This continues to prove that Episode 5 is the best episode in the saga so far because not only has most people seen Episode 5, Episode 5 is still ranked as the best star wars movie of all-time.

# Segmenting our data into seperate columns based on Gender.
Now it's time to examine how certain segments of the survey population responded. There are several columns that segment the data into two groups. Here are a few examples:

- Do you consider yourself to be a fan of the Star Wars film franchise? - True or False
- Do you consider yourself to be a fan of the Star Trek franchise? - Yes or No
- Gender - Male or Female

We can split a dataframe into two groups based on gender in the form of a binary column by creating two subsets of the gender column. The subsets will allow us to compute the most viewed movie, the highest-ranked movie, and other statistics separately for each group.

```python
import numpy as np

males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

males_seen = males[males.columns[3:9]].sum()
females_seen = females[females.columns[3:9]].sum()

N=6
ind = np.arange(N)
width = 0.25

fig, ax = plt.subplots(figsize=(12, 6))
rect1 = ax.bar(ind+width, males_seen, width, color='g')
rect2 = ax.bar(ind+width, females_seen, width, color='y')
ax.set_title("Star Wars Views By Gender Per Episode")
ax.set_ylabel("Number of Views")
ax.set_xlabel("Star Wars Episodes")
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(("Episode 1", "Episode 2", "Episode 3"
                    , "Episode 4", "Episode 5", "Episode 6"))

ax.legend((rect1[0], rect2[0]), ('Male', 'Female'), loc="lower right")
```

```python
male_rankings = males[males.columns[9:15]].mean()
female_rankings = females[females.columns[9:15]].mean()

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(ind, male_rankings, width, color='g')
ax.set_title("Star Wars Male Rankings Per Episode")
ax.set_ylabel("Ranking")
ax.set_xlabel("Star Wars Episodes")
ax.set_xticklabels(("Episode 1", "Episode 2", "Episode 3"
                    , "Episode 4", "Episode 5", "Episode 6"))

# for label in ax.xaxis.get_xticklabels():
#     label.set_horizontalalignment('right')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(ind, female_rankings, width, color='y')
ax.set_title("Star Wars Female Rankings Per Episode")
ax.set_ylabel("Ranking")
ax.set_xlabel("Star Wars Episodes")
ax.set_xticklabels(("Episode 1", "Episode 2", "Episode 3"
                    , "Episode 4", "Episode 5", "Episode 6"))
```

```python
print("Number of Male survey participants:\n", males_seen)
print("Number of Female survey participants:\n", females_seen)
```

What's interesting is that both male and female audiences ranked Star Wars Episode 5: The Empire Strikes Back as the greatest Star Wars movie of all time on average. What's also interesting is that Male audiences ranked Episode 4 higher than Episode 6 while Female Audiences had the exact opposite rankings. However, both Male and Female audiences had similar rankings for Episodes 1, 2 and 3 even though male audiences had higher average rankings than female audiences for each movie. This could be because there are significantly more male survey participants than female who seen each these movies.
