PART TWO

- Next we use our `explore_data` function to explore each of the variables that contain each of our datasets as lists of lists.
- To test it out we also returned a few columns from each dataset.

```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n')
        
    if rows_and_columns:
        print('Number of rows: ', len(dataset))
        print('Number of columns: ', len(dataset[0]))

file1 = open('AppleStore.csv')
from csv import reader
read_file1 = reader(file1)
iOS_data = list(read_file1)
iOS_header = iOS_data[0]
iOS_data = iOS_data[1:]

file2 = open('googleplaystore.csv')
from csv import reader
read_file2 = reader(file2)
android_data = list(read_file2)
android_header = android_data[0]
android_data = android_data[1:]

explore_data(iOS_data, 0, 6, True)
explore_data(android_data, 1, 6, True)
```

PART THREE
- One of the things we don't want in our datasets are duplicate apps. We start by creating empty lists to store duplicates for each of our datasets such as `iOS_dupicates` and `android_duplicates`.
- We also have to create a unique apps list for each dataset as well to keep track of all the unique apps in each of our large datasets. Our unique lists are `android_uniques` and `iOS_uniques`.
- To perform storing the duplicates and unique apps in the right lists that we create, we loop through each app and if they aren't stored in the 'uniques' lists already, then store them in the 'uniques' list. Otherwise store them in the duplicates list

```python
android_duplicates = []
iOS_duplicates = []
android_uniques = []
iOS_uniques = []

for app in android_data:
    name = app[0]
    if name in android_uniques:
        android_duplicates.append(name)
    else:
        android_uniques.append(name)
print('Android apps that have duplicates: ', android_duplicates)
print('Amount of Android duplicate apps: ', len(android_duplicates))

for app in iOS_data:
    name = app[1]
    if name in iOS_uniques:
        iOS_duplicates.append(name)
    else:
        iOS_uniques.append(name)
print('iOS apps that have duplicates: ', iOS_duplicates)
```

PART 5: Building a frequency table to find trending Apps

- We have to find the most profitable apps from the Android/ Google Play Data Set. To do that we have to take a few steps into consideration:
    - Creating the minimal version of the app from the Android Dataset
    - If the app gets good responses from its users than we develop the app even further.
    - If the app is profitable for 6 months, we create an iOS version of the app.
- Our goal is to find apps that will be successful in both markets. We start by looking at genres that are the most common in each market. We also need to build a frequency table for a few columns.
- We build a function that determines the frequency of certain columns in a table. We'll look at the prime_genre, genres, and category columns of the Google Play Data set.

```python
def find_frequency (dataset, index): #find_frequency function designed to find frequency for any dataset, and 
    #specific columns in that dataset
    freq_table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        
        if value in freq_table:
            freq_table[value] += 1 #If value is already there, increase frequency value by 1
        else:
            freq_table[value] = 1  #IF value not there, set value frequency to 1
            
    table_percentages = {}   #Dictionary storing the percentage of how much of these values appear in the dataset
    for key in freq_table:
        app_percentage = (freq_table[key] / total) * 100
        table_percentages[key] = app_percentage
   
    return table_percentages

def display_frequency(dataset, index): #Function to display the frequency
    table = find_frequency(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    sorted_table = sorted(table_display, reverse = True)
    for entry in sorted_table:
        print(entry[1], ':', entry[0])

display_frequency(iOS_final, -5)
```

- As you can see from the results pulled up from above, Gaming apps are the most common genre while Entertainment is the runner up.
- There's a huge gap between the most popular and second most popular genre. As you can see most people prefer to play games on their iPhones than anything else. 
- Social media accounts for only 3 percent of the popularity which makes sense considering there are only a few very popular Social Media apps that people visit daily. 
- The more practical genres are the most rare which shows that people only use those apps when they need to and there's only a handful of popular apps designed for practical use such as those for Health and Fitness, Food and Drink, and Medical.
- Most of the apps are designed for Entertainment than the practical uses. About 75% of all these apps shown belong in the Entertainment section.

Part 7
- Now we'll use our `find_frequency` function to find the frequency of the 'Category' column in our Google Play/ Android dataset. Just as we did for the 'genre' category in the App Store dataset.

```python
categories_android = find_frequency(android_final, 1)
num_apps = 0
for category in categories_android:
    total = 0
    len_category = 0
    hundred_k_plus = 0
    million_installs = 0
    five_mil_plus = 0
    for app in android_final:
        category_app = app[1]
        if category_app == category:
            n_installs = app[5]
            n_installs = n_installs.replace('+', '')
            n_installs = n_installs.replace(',', '')
            n_installs = float(n_installs)
            total += n_installs
            len_category += 1
            if n_installs >= 100000:
                hundred_k_plus += 1
            if n_installs == 1000000:
                million_installs += 1
            if n_installs >= 5000000:
                five_mil_plus += 1

    print('Amount of apps in the',category, 'category: ', len_category)
    print('Amount of apps with 100,000+ installs:', hundred_k_plus)
    print('Amount of apps with 1,000,000 installs:', million_installs)
    print('Amount of apps with 5,000,000+ installs:', five_mil_plus)
    print('Percentage of 100,000+ installs: ', ((hundred_k_plus / len_category) * 100),'%')
    print('Percentage of 1,000,000 installs: ', ((million_installs / len_category)* 100),'%')
    print('Percentage of 5,000,000+ installs: ', ((five_mil_plus / len_category) * 100),'%')
    avg_num_installs = total / len_category
    print (category, 'Average number of installations: ', avg_num_installs)
    print ('\n')
    
```
