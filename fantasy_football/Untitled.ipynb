```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

qb_games = pd.read_csv('qb_weekly_pass_stats.csv')

def team_abbrevs(team):
    str_len = len(team.split())
    if team == 'Jacksonville Jaguars':
        return 'JAX'
    elif team == 'New York Giants':
        return 'NYG'
    elif team == 'New York Jets':
        team = 'NYJ'
        return team
    elif team == 'Washington Football Team':
        return 'WAS'
    elif team == 'Los Angeles Rams':
        return 'LAR'
    elif team == 'Los Angeles Chargers':
        return 'LAC'
    else:
        if str_len == 2:
            return team[:3].upper()    
        else:
            team = team.split()[0:2]
            team_abbrev = ''
            for word in team:
                team_abbrev += word[0].upper()
            
            return team_abbrev


qb_games['Team'] = qb_games['Team'].apply(lambda x:team_abbrevs(x))
qb_games['Opponent'] = qb_games['Opponent'].apply(lambda x: team_abbrevs(x))
del(qb_games['Unnamed: 0'])
qb_games.drop([x for x in range(480, 524)], inplace=True)
qb_games['Week'] = qb_games['Week'].astype(int)
qb_games.columns = qb_games.columns.str.lower()
qb_games.columns = qb_games.columns.str.replace(' ', '_')
qb_games.rename(columns = {'player_name':'name'})
```
