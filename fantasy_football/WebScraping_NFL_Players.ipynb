```python
from requests import get
from bs4 import BeautifulSoup as bs

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_1_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'}

url = 'https://www.footballdb.com/players/current.html?letter=A'

response = get(url, headers=headers)
soup = bs(response.text, 'html.parser')
```

```python
player_containers = soup.find_all('div',{'class':'tr'})
print(type(player_containers))
print(len(player_containers))
```

```python
first_player = player_containers[1]
first_player
```

```python
first_player_data = first_player.find_all('div', {'class':'td'})
print(first_player_data[0].text)
print(first_player_data[1].text)
print(first_player_data[2].text)
print(first_player_data[3].text)
```

```python
#Lists that store every players attributes
names = []
teams = []
positions = []
colleges = []

#Appending the texts of each player attribute to their respective lists.
for container in player_containers[1:]:
    player = container.find_all('div',{'class':'td'})
    
    name = player[0].text
    names.append(name)
    
    position = player[1].text
    positions.append(position)
    
    team = player[2].text
    teams.append(team)
    
    college = player[3].text
    colleges.append(college)


print(len(names))
print(len(teams))
print(len(positions))
print(len(colleges))
```

```python
import string 
from random import randint
import time
start_time = time.time()
from IPython.core.display import clear_output
from warnings import warn
warn("Warning Simulation")

start_time = time.time()
requests = 0

#Lists containing all the clean texts for each player attribute.
names = []
teams = []
positions = []
colleges = []

# list of ever letter in the alphabet.
alphabet = list(string.ascii_lowercase)
urls = []

#for each letter in alphabet
for letter in alphabet:
    
#     response = get(
    urls.append('https://www.footballdb.com/players/current.html?letter='+letter)

for url in urls:
    response = get(url, headers=headers)
    # Pause the loop
    time.sleep(randint(8,15))

    # Monitor the requests
    requests += 1
    elapsed_time = time.time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)

    # Throw a warning for non-200 status codes
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))

    # Break the loop if the number of requests is greater than expected
    if requests > 72:
        warn('Number of requests was greater than expected.')
        break
        
    # Parse the content of the request with BeautifulSoup
    page_html = bs(response.text, 'html.parser')
        
    # Select all the player containers from a single page
    player_container = page_html.find_all('div', {'class':'tr'})
        
    for container in player_container[1:]:
        player = container.find_all('div',{'class':'td'})
    
        name = player[0].text
        names.append(name)

        position = player[1].text
        positions.append(position)
        
        team = player[2].text
        teams.append(team)

        college = player[3].text
        colleges.append(college)
        
```

```python
players_df = pd.DataFrame({'Player Name': names, 'Team': teams,
                        'Position':positions, 'College': colleges})
print(players_df.info())
```
